```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox as msgbox
import pandas as pd

df = pd.read_csv('GIG-logistics.csv')
departments = sorted(df['Department'].unique().tolist())

def verify():
    dept = dept_dropdown.get()
    name = name_entry.get()
    if not name or dept == "Select Department":
        msgbox.showinfo("Information Input Error.", "Please enter your name and select a department.")
        return
    match = df[(df['Name'].str.lower() == name.lower()) & (df['Department'] == dept)]
    if not match.empty:
        welcomeMessage(dept, name)
    else:
        msgbox.showinfo("Access Denied", "The name or department provided doesn't match our records.")

def welcomeMessage(dept, name):
    window = tk.Toplevel(root)
    window.title(f"{dept} Window")
    window.geometry("500x200")

    welcome = tk.Label(window, text=f"Welcome {name}! You are one of us at GIG Logistics! \n")
    welcome.pack()
    body = tk.Label(window, text=f"You have been verified, Click the button below to see all department members.")
    body.pack()

    button = tk.Button(window, text=f"Department Members", command=members)
    button.pack()
    button.config(fg="black", bg="yellow")
    
    root.mainloop()

def members():
    dept = dept_dropdown.get()
    deptmembers = df[df['Department'] == dept]['Name'].tolist()
    msgbox.showinfo("Department Members: \n", deptmembers)

root = tk.Tk()
root.title("GIG LOGISTICS STAFF VERIFICATION")
root.geometry("500x300")

name_label = tk.Label(root, text="Please Enter your Full Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

dept_label = tk.Label(root, text="Department:")
dept_label.pack()
dept_dropdown = ttk.Combobox(root, values=departments, state="readonly")
dept_dropdown.pack()
dept_dropdown.set("Select Department")

verify_button = tk.Button(root, text="Verify", command=verify)
verify_button.pack()

root.mainloop()
```
