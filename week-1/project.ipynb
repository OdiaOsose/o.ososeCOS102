```python
# Simple Interest
P = float(input("Enter Principal: "))
R = float(input("Enter Rate (%): "))
T = float(input("Enter Time Period (years): "))

A = P * (1+(R/100)*T)
I = A - P
print("Total Amount = ", A)
print ("Interest = ", I)

# <img src="simple interest.drawio.png">
```

```python
# Compound Interest
P = float(input("Enter Principal: "))
R = float(input("Enter Rate (%): "))
n = float(input("Enter number of times interest was compounded per year: "))
t = float(input("Enter Time Period (years): "))

r = R/100
A = P * (1 + r/n) ** (n*t)
I = A - P
print("Total Amount = ", A)
print("Interest = ", I)

# <img src="compound interest.drawio.png">
```

```python
# Annuity Plan
PMT = float(input("Enter Payment Amount: "))
R = float(input("Enter Rate (%): "))
n = int(input("Enter number of times interest was compounded per year: "))
t = float(input("Enter Time Period: "))

r = R/100
rate_pp = r/n
num = (((1+rate_pp)**(n*t)) - 1)
A = PMT * (num / rate_pp)
print("Total Amount = ", A)

# <img src="annuity plan.drawio.png">
```
