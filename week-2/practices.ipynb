```python
# practice 1a
str = input ("Enter your state of origin")
print ("Your state of origin is: ", str)
print ("The first character is: ", str[0])
print ("The 3rd to 5th characters are: ", str[2:5])
print ("Characters starting from the 3rd are: ", str[2:])
print ("State of origin two times", str * 2)
```

```python
# practice 1b
# Input from user
m = float(input("Enter mass in kg: "))

# Constant value for the speed of light in m/s
c = 299792458

# Calculating energy using Einstein's equation
energy = m * (c ** 2)

# Result
print (f"The energy equivalent to {m} kg of mass is {energy} joules.")
```

```python
# practice 1c
list = [ 'Anaconda', 786, 2.23, 'Jupyter', 70.2 ]
shortlist = [321, 'Python']

print(list)              # Prints complete list
print(list[0])           # Prints first element of the list
print(list[1:3])         # Prints elements starting from 2nd till 3rd
print(list[2:])          # Prints elements starting from the 3rd element
print(shortlist * 2)     # Prints list two times
print(list + shortlist)  # Prints concatenated lists
```

```python
# practice 1d
tuple = ("Ekiti", 750, 'Oshogbo', 250, "Akure", 500)
s_tuple = ("Abeokuta", 300, "Ogbomosho")

# Prints the complete tuple
print (tuple)

# Prints last element of the tuple 
print (tuple[-1])

# Prints elements of the tuple starting from 2nd till 3rd
print (tuple[1:3])

# Prints elements of the tuple starting from 3rd element
print (tuple[3:])

# Prints the contents of the tuple thrice
print (s_tuple *3)

# Prints concatenated tuples
print (tuple + s_tuple)
```

```python
# practice 2a
# Python program to swap two cities

# To take inputs from the user
city_1 = input('Enter name of City 1: ')
city_2 = input('Enter name of City 2: ')

# create a temporary variable and swap the cities
temp = city_1
city_1 = city_2
city_2 = temp

print(f"The name of City 1 after swapping is {city_1}")
print(f"The name of City 2 after swapping is {city_2}")
```

```python
# Program to check if a number is positive, negative or 0

num = float(input("Enter a number: "))
if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```
