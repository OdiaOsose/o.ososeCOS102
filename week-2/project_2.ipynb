```python
import numpy as np
def findroots():
    print ("Polynomial Equation Roots Finder")
    print ("Equation type options: ")
    print("1. Quadratic (Ax² + Bx + C = 0)")
    print("2. Cubic (Ax³ + Bx² + Cx + D = 0)")
    print("3. Quartic (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
    choice = input("Choose an equation type (1-3): ")

    if choice == "1":
        print("Enter coefficients for Ax² + Bx + C = 0: ")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        coeffs = [a, b, c]

    elif choice == '2':
        print("Enter coefficients for Ax³ + Bx² + Cx + D = 0")
        a = float(input("A: "))
        b = float(input("B: "))
        c = float(input("C: "))
        d = float(input("D: "))
        coeffs = [a, b, c, d]
        
    elif choice == '3':
        print("Enter coefficients for Ax⁴ + Bx³ + Cx² + Dx + E = 0")
        a = float(input("A: "))
        b = float(input("B: "))
        c = float(input("C: "))
        d = float(input("D: "))
        e = float(input("E: "))
        coeffs = [a, b, c, d, e]
    
    else:
        print("Invalid coefficient selection.")
        return

    # Using numpy to find roots
    roots = np.roots(coeffs)

    print ("The roots of the equation are: ")
    for r in roots:
        print(np.round (r,4))

findroots()
```
