# 1st task

BEGIN
SET TOTAL = 0
REPEAT INPUT ItemPrice
COMPUTE Total = Total + ItemPrice
UNTIL No more items

OUTPUT Total
INPUT CashReceived
IF CashReceived >= Total
   COMPUTE Change = CashReceived - Total
   OUTPUT Change
   PRINT Receipt
ELSE PRINT Insufficient Funds
END
